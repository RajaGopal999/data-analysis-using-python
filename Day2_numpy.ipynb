```python
#joining arrays
a=np.array([2,6,9,72,19,3]).reshape(2,3)
b=np.array([3,4,7,2,1,4]).reshape(2,3)
print(a)
print("_____________")
print(b)
print("________________")
print("After vertically stacking arrays!")
print(np.vstack((a,b)))
```

```python
a=np.array([2,6,9,72,19,3]).reshape(2,3)
b=np.array([3,4,7,2,1,4]).reshape(2,3)
print(a)
print("_____________")
print(b)
print("________________")
print("After horizontally stacking arrays!")
print(np.hstack((a,b)))
```

```python
a=np.arange(30).reshape(2,3,5)#original -->(2,3,5)  in dstack-->(3,5,2)
print(a)
print("after dstack")
print(np.dstack(a))#no.of rows --> no of groups
#no.of col in every group-->no.of rows in every gruop
#[0 0],[1 1],....
```

Statistics

```python
#mean,median,variance,standard dev
a=np.array([12,34,25,66,66,71,81])
print("Mean",np.mean(a))
print("Median ",np.median(a))
print("variance ",np.var(a))
print("Standard Dev ",np.std(a))
```
