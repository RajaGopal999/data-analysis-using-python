```python
'''there are 5 subjects psychology,nano technology,medicine,onchology,anthropology.two students have applied to do masters in an of these 5 fields with the line plot
show in which field they can excel the cutoff for each subject is 70 .'''
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cut=[70,70,70,70,70]
sub=['psy','nano','medi','oncho','anthro']
#plt.plottype(x,y,label,color,marker,linewidth,linestyle)
plt.scatter(sub,s1,label="student1",color="k",marker="P")#student1
plt.scatter(sub,s2,label="student2",color="m",marker="H")#student2
plt.plot(cut,label="cutoff",color="y",linestyle=":",linewidth=4)#cutoff
plt.legend()#box is created
plt.title("best subject")
plt.ylabel(["subject"])
plt.xlabel("marks")
```

```python
'''an india vs australia is match on live india has given 298 as target gor australia the runs for every 5 over is given as follow 25,51,84,131,160,189,220,250,267,297
australia have complited 25 overs 15,41,94,110,151
plot overs vs runs for both the countries with india has a blue line and aus have yellow line give appropriate labels and legend the marker for india is flag and aus is plus
take over arange funtion create runs as array linewidth=2'''
import numpy as np
i=np.array([25,51,84,131,160,189,220,250,267,297])
a=np.array([15,41,94,110,151])
over=np.arange(0,50,5)
over1=np.arange(0,25,5)
plt.plot(over,i,label="india",color="blue",marker=">")
plt.plot(over1,a,label="australia",color="yellow",marker="P")
plt.legend()
plt.grid()
plt.title("ind vs aus")
plt.ylabel("scores")
plt.xlabel("overs")
```
