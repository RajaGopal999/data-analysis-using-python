```python
#sets
s={1,2,3,4}
print(type(s))#it is a type of set
print(s)#indexes are always changes
s1={1,2,36,36,91}#no duplicate values
print(s1)
l=["hii","here"]
s2=set(l)#type conversion
print(s2)
#add values
s3={1,2,3}
s3.add(4)
print(s3)
#for inserting multiple data
s4={3,6,9}
s4.update([1,2,3])
print(s4)
#removing an element
s4.discard(2)
print(s4)
a={1,2,3,4,6}
b={1,2,6,9}
print(a.union(b))#union
print(a.intersection(b))#intersection
print(a.difference(b))#difference
print(a.symmetric_difference(b))#symmetric difference
print(a.issuperset(b))
print(a.issubset(b))
print(a.isdisjoint(b))
```
